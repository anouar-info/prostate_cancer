```python
clinical_df = pd.read_csv("./prostate_cancer/data_clinical_patient.txt",sep="\t",comment="#")
print(clinical_df.columns)
clinical_df.head()
```

```python
mutations_df = pd.read_csv("./prostate_cancer/data_mutations.txt",sep="\t",comment="#")
print(np.array([e for e in mutations_df.columns]))
mutations_df.head()
```

```python
mutations_df[['Hugo_Symbol','Tumor_Sample_Barcode']].head(10)
```

```python
top_20_genes = mutations_df.value_counts("Hugo_Symbol")[:20]
top_10_genes = mutations_df.value_counts("Hugo_Symbol")[:10]

top_20_genes
```

```python
oncogenes = [e for e in top_10_genes.index]

mutation_matrix = pd.crosstab(mutations_df["Tumor_Sample_Barcode"] , mutations_df["Hugo_Symbol"])

mutation_matrix = mutation_matrix[mutation_matrix.columns.intersection(oncogenes)]

mutation_matrix.sample(10)
```

```python
mutation_matrix.index = mutation_matrix.index.str.slice(0,9)

final_df = mutation_matrix.join(clinical_df.set_index("PATIENT_ID"), how = "inner")

final_df.sample(10)
```
